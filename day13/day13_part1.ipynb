```python
import numpy as np


with open("input.txt") as f:
    lines = f.readlines()
    blocus = []
    curr = []
    for line in lines:
        if line == '\n':
            blocus.append(np.array(curr))
            curr = []
        else:
            curr.append(list(line.strip('\n')))
    blocus.append(np.array(curr))
blocus
```

```python
som = 0
for arr in blocus:
    row_idx = 0
    col_idx = 0
    n, m = arr.shape
    for i in range(n-1):
        j = 1
        while i-j+1 >= 0 and i+j < n and (arr[i-j+1] == arr[i+j]).all():
            j+=1

            if i-j+1 < 0 or i+j >= n:
                row_idx = i+1
                break

    for i in range(m-1):
        j = 1
        while i-j+1 >= 0 and i+j < m and (arr[:,i-j+1] == arr[:, i+j]).all():
            j+=1

            if i-j+1 < 0 or i+j >= m:
                col_idx = i+1
                break
    print(row_idx, col_idx)
    som += col_idx + 100*row_idx

som
```
