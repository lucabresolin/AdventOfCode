```python
import sys
sys.setrecursionlimit(4000)

with open("input.txt") as f:
    lines = tuple([line.strip('\n') for line in f.readlines()])
lines
```

```python
def get_new_direction(current_direction, symbol):
    x, y = current_direction
    if symbol == '.' or (x==0 and symbol == '|') or (y==0 and symbol == '-'):
        return [current_direction]
    elif symbol == '-' and y!=0 :
        return [(1,0),(-1,0)]
    elif symbol == '|' and x!=0:
        return [(0,1),(0,-1)]
    elif symbol == '/' :
        return [(-current_direction[1], -current_direction[0])]
    elif symbol == '\\' :
        return [(current_direction[1], current_direction[0])]
    else:
        print("PIPOU")

def inbound(coord):
    x, y = coord
    return x>=0 and x<len(lines) and y>=0 and y<len(lines[0])

def move(coord, dir):
    x, y = coord
    u, v = dir
    return (x+u, y+v)

def print_grid(viz):
    lop = []
    for i in range(len(lines)):
        lop.append("")
        for j in range(len(lines[0])):
            if (j,i) in viz:
                lop[-1] = lop[-1] + '#'
            else:
                lop[-1] = lop[-1] + '.'
        print(lop[-1 ])

def run(current_dir, current_pos):
    visited_spot = set([])
    visited_spot_dir = set([])

    def run_sub(current_dir, current_pos):
        if (current_pos,current_dir) not in visited_spot_dir:
            visited_spot.add(current_pos)
            visited_spot_dir.add((current_pos,current_dir))

            x, y = current_pos
            symbol = lines[y][x]
            next_dir = get_new_direction(current_dir, symbol)
            
            for dir in next_dir:
                next_pos = move(current_pos, dir)
                if inbound(next_pos):
                    run_sub(dir, next_pos)

    run_sub(current_dir, current_pos)
    # print(current_dir, current_pos)
    # print_grid(visited_spot)
    # print(len(visited_spot))
    # print()
    return len(visited_spot)
```

```python
max = 0
max_coord = None
to_explore = []

for idle_term in (0, len(lines)-1):
    for i in range(len(lines[0])):
        starting_point = (i, idle_term)
        starting_direction = (0, 1 if idle_term==0 else -1)        
        to_explore.append((starting_direction, starting_point))
for idle_term in (0, len(lines[0])-1):
    for i in range(len(lines)):
        starting_point = (idle_term, i)
        starting_direction = (1 if idle_term==0 else -1, 0)
        to_explore.append((starting_direction, starting_point))
    


for current_dir, current_pos in to_explore:
    energy = run(current_dir, current_pos)
    if energy > max:
        max = energy
        max_coord = starting_point

max
```
