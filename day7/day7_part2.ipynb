```python
with open("input.txt",'r') as f :
    lines = f.readlines()
lines
```

```python
cards = [elem.split()[0] for elem in lines]
bids = [int(elem.split()[1].replace('\n','')) for elem in lines]
print(cards, bids)
```

```python
def distinct_char(text):
    end = []
    for char in text:
        if char not in end and char != 'J':
            end.append(char)
    return len(end)

def has_pair(card):
    card = card.replace('J','')
    occ = {}
    for char in card:
        if char in occ:
            occ[char] +=1
        else:
            occ[char] = 1
    return 2 in occ.values()

def is_five(card):
    return distinct_char(card)<=1

def is_four(card):
    return distinct_char(card)<=2 and (not has_pair(card) or card.count('J')>=2)

def is_full(card):
    return distinct_char(card)<=2 and has_pair(card)

def is_toak(card):
    if card.count('J')>=2:
        return True

    occ = {}
    for char in card:
        if char in occ:
            occ[char] +=1
        else:
            occ[char] = 1
    return (3 in occ.values() and len(occ.keys())==3) or (2 in occ.values() and len(occ.keys())==4 and 'J' in occ.keys())

def two_pair(card):
    return distinct_char(card)<=3 and has_pair(card)

def one_pair(card):
    return distinct_char(card)<=4 


def get_points(card):
    if is_five(card):
        return 500
    elif is_four(card):
        return 400
    elif is_full(card):
        return 300
    elif is_toak(card):
        return 250
    elif two_pair(card):
        return 100
    elif one_pair(card):
        return 50
    else:
        return 10

symbol_convert = {
        "A":14, "K":13, "Q" : 12, "J":1, "T" : 10
    }

def get_numeric_card(sym):
    if not sym.isnumeric():
        sym = symbol_convert[sym]
    return int(sym)


def get_highest_card_point(card):
    max_actual = 0
    for symb in card:
        if get_numeric_card(symb)>max_actual:
            max_actual = get_numeric_card(symb)
    return max_actual

def highest_card(sym1, sym2):
    return get_numeric_card(sym1) < get_numeric_card(sym2)
```

```python
games_ordered = []

def place_game(local_game, local_point, local_bid):
    if len(games_ordered)==0:
        return [(local_game, local_point, local_bid)]

    for idx, g_and_p in enumerate(games_ordered):
        comp_game, comp_points, _ = g_and_p
        if local_point < comp_points:
            return games_ordered[:idx] + [(local_game, local_point, local_bid)] + games_ordered[idx:]
        elif local_point == comp_points:
            for char1, char2 in zip(local_game, comp_game):
                if highest_card(char1, char2):
                    return games_ordered[:idx] + [(local_game, local_point, local_bid)] + games_ordered[idx:]
                if highest_card(char2, char1):
                    break

    return games_ordered + [(local_game, local_point, local_bid)]

for current_game, current_bid in zip(cards, bids):
    games_ordered = place_game(current_game, get_points(current_game), current_bid)
games_ordered
```

```python
sums = 0
for idx, val in enumerate(games_ordered):
    sums += val[2] * (idx+1)
sums
```
