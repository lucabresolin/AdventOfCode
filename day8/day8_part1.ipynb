```python
with open("input.txt", 'r') as f :
    lines = [line.strip('\n').replace(" ","") for line in f.readlines() if line != '\n']
lines
```

```python
instructions = lines[0]

dicoton = {}
for chemin in lines[1:]:
    cle, choix = chemin.split("=")
    gauche, droite = choix.strip('()').split(',')
    dicoton[cle] = {'L' : gauche, 'R': droite}
dicoton
```

```python
endroit = 'AAA'
n = 0
while endroit != 'ZZZ':
    endroit = dicoton[endroit][instructions[n%len(instructions)]]
    n+=1
n
```
