```python
with open('input.txt') as f:
    lines = [line.strip('\n') for line in f.readlines()]
lines

coord = (-1, -1)
for i, c in enumerate(lines[0]):
    if c=='.':
        coord = (0, i)
coord, lines
```

```python
def inbound(x, y):
    return x in range(len(lines)) and y in range(len(lines[0]))

dir_symbol = {'>':(0,1), '<':(0,-1), '^':(-1, 0), 'v':(1,0)}

already_seen = ()

def valid_dir(symbol, direction):
    is_point = symbol == '.'
    is_valid_dir = dir_symbol[symbol] == direction if symbol in dir_symbol else False
    return is_point or is_valid_dir

def run(x, y, stride_length, px, py):
    if x==len(lines)-1 and lines[x][y]=='.':
        return stride_length
    
    max_run = 0
    for u, v in ((1,0),(0,1),(-1,0),(0, -1)):
        next_x, next_y = x+u, y+v
        if inbound(next_x, next_y) and not(next_x==px and next_y==py) and valid_dir(lines[x][y], (u,v)):
            current_run = run(next_x, next_y, stride_length+1, x, y)
            if current_run > max_run:
                max_run = current_run

    return max_run
run(*coord, 0, 0,0)
```
