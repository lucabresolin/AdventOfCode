```python
with open('input.txt') as f:
    lines = [line.strip('\n') for line in f.readlines()]
n = len(lines)
start_coord = (0, 1)
end_coord = (n-1, n-2)
lines
```

```python
def inbound(x, y):
    return x in range(n) and y in range(n)

carrefour = [start_coord, end_coord]
for x in range(n):
    for y in range(n):
        next_directions = [(x+a,y+b) for a, b in ((1,0),(0,1),(-1,0),(0, -1)) if inbound(x+a, y+b) and lines[x+a][y+b] != '#']
        if len(next_directions) >= 3:
            carrefour.append((x, y))
len(carrefour), carrefour
```

```python
distances = {pt : {} for pt in carrefour}

for inter in carrefour:
    seen = {inter}
    a, b = inter
    node_to_look = [(a, b, 0)]
    while len(node_to_look)>0:
        cx, cy, d = node_to_look.pop()
        if d>0 and (cx, cy) in carrefour:
            distances[(a,b)][(cx, cy)] = d
        else:
            for u, v in ((cx+1,cy),(cx,cy+1),(cx-1,cy),(cx, cy-1)):
                if inbound(u, v) and lines[u][v] != '#' and (u, v) not in seen:
                    node_to_look.append((u, v, d+1))
                    seen.add((u, v))


distances
```

```python
already_seen = ()

def run(x, y, stride_length, seen):
    if (x, y) in seen:
        return 0

    new_seen = seen.copy()
    new_seen.add((x, y))

    if x==n-1 and lines[x][y] == '.':
        return stride_length
        
    max_run = 0
    for u, v in distances[(x,y)]:
        current_run = run(u, v, stride_length+distances[(x,y)][(u,v)], new_seen)
        if current_run > max_run:
            max_run = current_run

    return max_run

run(*start_coord, 0, set())
```
