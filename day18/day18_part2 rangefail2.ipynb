```python
with open("input.txt") as file:
    instructions = [line.split()[2].strip('()#') for line in file.readlines()]
instructions
```

```python
charhexmap = {'A':10, 'B':11, 'C':12, 'D':13, 'E':14, 'F':15}
def get_hex_int(hex_char):
    if hex_char.isnumeric():
        return int(hex_char)
    else:
        return charhexmap[hex_char.upper()]

def get_hex_number(hexstr):
    nombor = 0
    for idx,c in enumerate(hexstr[::-1]):
        nombor += get_hex_int(c) * 16 ** idx
    return nombor

hexdirmap = {'0':'R', '1':'D', '2':'L', '3':'U'}

def decode_instr(hex_code):
    digit = get_hex_number(hex_code[:-1])
    dir = hexdirmap[hex_code[-1]]
    return digit, dir

instructions = [decode_instr(instr) for instr in instructions]
instructions
```

```python
dir_map = {
    'R':(0, 1),
    'L':(0, -1),
    'D':(1, 0),
    'U':(-1, 0)
}

def ordered_range(x, y):
    if x < y:
        return range(x, y+1)
    else:
        return range(y, x+1)

def get_edge_tile(instr):
    x, y = 0, 0
    edge_tiles = []
    minx = 0
    miny=0
    maxx = 0
    maxy=0
    for length, dir in instr:
        u, v = dir_map[dir]
        nx, ny = x+(u*length), y+(v*length)
        edge_tiles.append((ordered_range(x, nx), ordered_range(y, ny)))

        x, y = nx, ny
        minx = minx if minx<x else x
        miny = miny if miny<y else y
        maxx = maxx if maxx>x else x
        maxy = maxy if maxy>y else y

    return edge_tiles, minx, miny, maxx, maxy

tiled, xmin, ymin, xmax, ymax = get_edge_tile(instructions)
W = ymax+1
H = xmax+1
tiled = [(range(x.start-xmin, x.stop-xmin), range(y.start-ymin, y.stop-ymin)) for x, y in tiled]
len(tiled)
```

```python
def new_range(*args):
    max_val = max(args)
    min_val = min(args)
    return range(min_val, max_val)

def range_contiguous(ranga, rangb):
    a, b = ranga.start, ranga.stop
    c, d = rangb.start, rangb.stop
    return a == d-1 or b-1 == c

def range_included(ranga, rangb):
    a, b = ranga.start, ranga.stop
    c, d = rangb.start, rangb.stop
    return a >=c and b<=d

def get_line_length(row_num):
    edges = []
    for xran, yran in tiled[::-1]:
        if row_num in xran:
            # print(yran)
            # print(edges)
            if len(yran)==1:
                if len(edges)>0 :
                    if isinstance(edges[-1], int) or yran.start not in edges[-1]:
                        edges.append(yran.start)
                    
                else:
                    edges.append(yran.start)
            else:
                if len(edges)>0:
                    prev = edges[-1]
                    if isinstance(prev, int) and prev in yran:
                        edges[-1] = new_range(yran.start, yran.stop, prev, prev+1)
                    elif range_contiguous(yran, prev):
                        edges[-1] = new_range(prev.start, prev.stop, yran.start, yran.stop)
                    elif not range_included(yran, prev):
                        edges.append(yran)

                elif yran not in edges:
                    edges.append(yran)
    
    out = []
    for idx in range(0, len(edges)-1, 2):
        curr = edges[idx]
        next = edges[idx+1]
        if isinstance(curr, int):
            if isinstance(next, int):
                out.append(ordered_range(curr, next))
            else:
                out.append(new_range(curr, curr+1, next.start, next.stop))
        else:
            if isinstance(next, int):
                out.append(new_range(next, next+1, curr.start, curr.stop))
            else:
                out.append(curr)
                out.append(next)

        
    return out

```

```python
def calculate_tile(pere):
    som = 0
    for elem in pere:
        som += len(elem)
    return som
```
