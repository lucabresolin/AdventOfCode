```python
with open('input.txt' ,'r') as f:
    lines = [[int(sl) for sl in line.strip('\n').split()] for line in f.readlines()]

lines
```

```python
def not_all_zero(lis):
    for e in lis:
        if e != 0:
            return True
    return False

def next_sublist(lis):
    res = []
    for idx in range(1, len(lis)):
        res.append(lis[idx]-lis[idx-1])
    return res

def get_next_digit(serie):
    serie_degrad = [serie]
    for s in serie_degrad:
        if not_all_zero(s):
            serie_degrad.append(next_sublist(s))

    next_digit = serie[-1]
    for sl in serie_degrad[1:]:
        next_digit += sl[-1]

    return next_digit
```

```python
sum([get_next_digit(line) for line in lines])
```
