```python
with open('input.txt' ,'r') as f:
    lines = [[int(sl) for sl in line.strip('\n').split()] for line in f.readlines()]

lines
```

```python
def not_all_zero(lis):
    for e in lis:
        if e != 0:
            return True
    return False

def next_sublist(lis):
    res = []
    for idx in range(1, len(lis)):
        res.append(lis[idx]-lis[idx-1])
    return res

def get_subs(serie):
    serie_degrad = [serie]
    for s in serie_degrad:
        if not_all_zero(s):
            serie_degrad.append(next_sublist(s))

    print(serie_degrad)
    return serie_degrad[:-1]

def get_next_digit(sublists):
    if sublists is None:
        return 0
    elif len(sublists) == 1:
        return sublists[0][0]
    else:
        return sublists[0][0] - get_next_digit(sublists[1:])
```

```python
sum([get_next_digit(get_subs(line)) for line in lines])
```
