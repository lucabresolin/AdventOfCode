```python
import numpy as np
with open("input.txt", 'r') as f:
    lines = np.array([list(line.strip('\n')) for line in f.readlines()])
lines
```

```python
def rollup_andfill(input):
    liste_dot = []
    liste_nondot = []
    liste_resultat = []
    
    for elem in input:
        if elem =='.':
            liste_dot.append('.')
        elif elem == '#':
            liste_resultat +=  liste_nondot + liste_dot + ['#']
            liste_dot = []
            liste_nondot = []
        else:
            liste_nondot.append(elem)

    liste_resultat += liste_nondot + liste_dot 
    return liste_resultat

from tqdm import tqdm

all_arr = []
for i in range(1000):
    for col in range(lines.shape[1]):
        lines[:, col] = rollup_andfill(lines[:, col])
        
    for row in range(lines.shape[0]):
        lines[row] = rollup_andfill(lines[row])

    for col in range(lines.shape[1]):
        lines[:, col] = rollup_andfill(lines[:, col][::-1])[::-1]

    for row in range(lines.shape[0]):
        lines[row] = rollup_andfill(lines[row][::-1])[::-1]

    a = [ (elme == lines).all() for elme in all_arr]
    print(a)
    if True not in a:
        all_arr.append(lines.copy())

    
len(all_arr)
```
