```python
workflows = {}

def superior_than(a, b):
    return a > b

with open("input.txt") as f:
    for line in f.readlines():
        if len(line)>1 and line[0] != '{':
            name, workflow = line.split('{')
            workflow = workflow.strip('{}\n').split(',')
            workflows[name] = ([], workflow.pop())
            for rule in workflow:
                condition, result = rule.split(':')
                term = condition[0]
                comparator = condition[1]
                limit = int(condition[2:])
                workflows[name][0].append((term,comparator,limit,result))
            


workflows
```

```python
def count(state, name):
    if name == 'A':
        prod = 1
        for mn, mx in state.values():
            prod *= mx - mn + 1
        return prod
    if name == 'R':
        return 0
    
    rules, default = workflows[name]
    total = 0
    for term, comparator, limit, result in rules:
        mn, mx = state[term]
        if comparator == '<':
            good_range = (mn, min(mx, limit-1))
            bad_range = (good_range[1]+1, mx)
        else:
            good_range = (max(mn, limit+1), mx)
            bad_range = (mn, good_range[0]-1)
        
        if good_range[1]>= good_range[0]:
            new_state = dict(state)
            new_state[term] = good_range
            total += count(new_state, result)
        
        if bad_range[1] >= bad_range[0]:
            state[term] = bad_range
        
        else:
            break
    else:
        total += count(state, default)

    return total


count({term : (1,4000) for term in 'xmas'}, "in")
```
