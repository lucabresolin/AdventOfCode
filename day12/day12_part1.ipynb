```python
with open("input.txt") as f :
    lines = [line.strip('\n').split() for line in f.readlines()]
lines
```

```python
def all_arangements(format):
    if format=="" or '?' not in format:
        return [format]
    else:
        next_arrang = all_arangements(format[1:])
        if format[0] == '?':
            return ['#'+arr for arr in next_arrang] + ['.'+arr for arr in next_arrang]
        else:
            return [format[0]+arr for arr in next_arrang]

def get_schema(arr):
    return ",".join([str(len(l)) for l in arr.split('.') if len(l)>0])


[ (arr, get_schema(arr)) for arr in  all_arangements(lines[0][0])]
```

```python
som = 0
for pattern, schema in lines:
    all_arra = all_arangements(pattern)
    n_valid = sum([get_schema(arr)==schema for arr in all_arra])
    som+=n_valid
    print(n_valid)
som
```
