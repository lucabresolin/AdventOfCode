```python
with open('input.txt') as f:
    lines = [list(map(lambda txt : [int(elem) for elem in txt.split(',')],line.strip('\n').split('~'))) for line in f.readlines()]
lines

lines = sorted(sorted(sorted(lines, key= lambda x : x[0][0]), key= lambda x : x[0][1]), key= lambda x : x[0][2])
```

```python
def overlap(brick1, brick2):
    start_coord1, end_coord1 = brick1
    start_coord2, end_coord2 = brick2

    x11, y11, z11 = start_coord1
    x12, y12, z12 = end_coord1
    x21, y21, z21 = start_coord2
    x22, y22, z22 = end_coord2

    for x in range(x11, x12+1):
        for y in range(y11, y12+1):
            for z in range(z11, z12+1):
                if x in range(x21, x22+1) and y in range(y21, y22+1) and z in range(z21, z22+1):
                    return True
    return False

def brickfall(brick):
    brick[1][2] = brick[1][2] - 1
    brick[0][2] = brick[0][2] - 1
    return brick

def brickclimb(brick):
    brick[1][2] = brick[1][2] + 1
    brick[0][2] = brick[0][2] + 1
    return brick

def overlap_any(brick, idx_avoid, wall = lines):
    overlap_list = [] 
    for idx_brick, curr_brick in enumerate(wall):
        if idx_avoid!= idx_brick and overlap(brick, curr_brick):
            overlap_list.append(idx_brick)
    return overlap_list
```

```python
support = {}
from tqdm import tqdm

for idx, current_brick in tqdm(enumerate(lines)):
    #print(current_brick)
    base_height = current_brick[0][2]
    for i in range(base_height-1):
        brickfall(current_brick)
        if current_brick[0][2] <= 0:
            current_brick[0][2] = base_height
            current_brick[1][2] += base_height
            break

        overlaps = overlap_any(current_brick, idx)
        if len(overlaps)>0:
            brickclimb(current_brick)
            support[idx] = overlaps
            break
    #print(current_brick)
    #print("====")
```

```python
print(support)
comptbyval = {key:0 for key in range(len(lines))}
for idx, vals in support.items():
    if len(vals) == 1:
        comptbyval[vals[0]] = 1
comptbyval
```
