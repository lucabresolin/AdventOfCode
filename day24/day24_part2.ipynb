```python
import sympy
with open('input.txt') as f:
    lines = list(map(lambda x : [[int(e) for e in l.split(', ')] for l in x.strip('\n').split(' @ ')]  ,f.readlines()))
lines   
```

```python
a, b, c, l, m, n = sympy.symbols("a, b, c, l, m, n")
equations = []

for coord, dir in lines:
    x, y, z = coord
    u, v, w = dir
    equations.append((a - x) * (v - m) - (b - y) * (u - l))
    equations.append((b - y) * (w - n) - (c - z) * (v - m))

answers = sympy.solve(equations)
sum(list(answers[0].values())[:3])
```
